**What kind of data curation techniques are required?**
    
    1) Delete the record which has missing values
    
        If we got most of the record as NaN data value then we can delete 
        that record, But we follow this step when we have large dataset then 
        it is advisable to that.Since in our dataset we have a quite huge
        dataset then we can delete the NaN value rows from 
        'name' attributes, But if we have small dataset then we should avoid
        this step, as each and every record of dataset specify a huge 
        amount of information, and that help the model to 
        understand learn more.
        
    2) Create a seperate model to handle the missing values
            
    
    3) Use stastical methods like (Mean, medium, Mode)
    
        Mean >> Simple Average of all the values present in that particular
        attribute
        
        Medium >> Sort the all values in the particular attribute in 
        ascending order and take middle value of that and replace it with 
        NaN value.
    
        Mode >> Whichever value is having highest frequency in that 
        attribute we will replace it with the NaN value.
